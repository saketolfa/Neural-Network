```python
# import libraries
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate a datset
x = np.array([-1.0 , 0.0, 1.0 ,2.0,3.0 ,4.0] ,dtype=float)# input(or features)
y= np.array([-3.0 , -1.0, 1.0 ,3.0,5.0 ,7.0] , dtype=float)# output(or target)

# dataset visualisation
plt.scatter(x , y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
#define the neural network with one layer , one neural and one input value
model  = keras.Sequential([keras.layers.Dense(units=1 , input_shape=[1])])
model.compile(optimizer = 'sgd' , loss= 'mean_squared_error',metrics=[tf.keras.metrics.Accuracy()])

    #Dense :  represents  the layer C ,
    #units :  number of neural in layer C,
    #input_shape : number of features. in our case we have just one feature, x.
    

#training model 500 times 
history = model.fit( x , y , epochs = 500)
```

```python
history.history.keys()
```

```python
print(len(history.history['loss']))

plt.plot(history.history['loss'] , label='loss _curve')
plt.xlabel('Iteration number')
plt.ylabel('loss-score')
plt.legend()
```

```python
# predict new value
print(model.predict([10.0]))
```
